# Zip-code wise Demographic and geospatial features for Udupi district

```python
df=pd.read_csv("Outscraper + census Data.csv")
df
```

# Zip-code wise aggregation of no of CVD patients in a hospital during the course of 2022

```python
card=pd.read_excel("Card_List1.xlsx")
card
```

# Merged dataset

```python
df=pd.merge(df, card, on="Postal Code")
```

```python
#handling missing values
import numpy as np

df=df.sort_values("Card number", ascending=True)
df['Hectares']=df['Hectares'].fillna(round(df['Hectares'].mean(), 2))
df['M-Female Cultivators']=df['M-Female Cultivators'].fillna(round(df['M-Female Cultivators'].mean(), 2))

df
```

```python
Y=df[['Card number']]
```

```python
X2=df[['Playgrounds', 'Dentists','M-Other Male Workers',
       'Literate Males', 'Drinking Bars' ]]


X=df[['Hectares','Households','Total population','Total Males','Total Females',
'Total(0-6)','Total males(0-6)','Total Females(0-6)','Total Workers','Total Male Workers',
'Total Female Workers','Non Workers','Non Male Workers','Non Female Workers','Literates',
 'Literate Males','Literate Females','Illiterates','Illiterate Males','Illiterate Females',
 'Main Workers','Main male workers','Main Female workers','I-Marginal Workers','I-Marginal Male',
 'I-Marginal Female','I-Agricultural Labourers','I-Agricultural Female','I-Agricultural Male','M-Agricultural Labourers',
 'M-Male Agricultural','M-Female Agricultural','I-Cultivators','I-Cultivators Male','I-Cultivators Female',
 'M-Cultivators','M-Male Cultivators','M-Female Cultivators','M-Household Industry Workers',
 'M-Household Male','M-Household Female','MG-Household Industry Workers','MG-Household Male',
 'MG-Household Female','M-Other Workers','M-Other Male Workers','M-Other Female Workers',
 'MG-Other Workers','MG-Other Male Workers','MG-Other Female Workers','Ayurveda','Banks',
 'Book Stores','Bus Stops','Colleges','Dentists','Drinking Bars','Fast Food And Caf','Food Markets',
 'Garden & Orchard','Gyms','Hospitals','Libraries','Mnt No','Museums','Natural Reserves','Opthalmologists',
'Parks','Pharmacies','Playgrounds','Restaurants','Schools','Sports Clubs','Temples']] 
```

# Decision tree regressor

```python
# import the regressor
from sklearn.tree import DecisionTreeRegressor 
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3,
                                                    random_state=112)
# create a regressor object
regressor = DecisionTreeRegressor(random_state = 0) 
  
# fit the regressor with X and Y data
regressor.fit(X, Y)
```

```python
# import export_graphviz
from sklearn.tree import export_graphviz 
import graphviz
import matplotlib.pyplot as plt
from sklearn import tree

  

dot_data=export_graphviz(regressor, out_file =None,
               feature_names =X.columns.tolist()) 

# Draw graph

graph = graphviz.Source(dot_data, format="png") 
graph.render("decision_tree_graphivz")
graph
```

# Feature importance attributes

# Split Count

```python
from xgboost import XGBRegressor
xgb = XGBRegressor().fit(X, Y["Card number"].tolist())
f = pd.Series(xgb.get_booster().get_score(importance_type='weight'))
split_count = f / f.sum() * 100

split_count=pd.DataFrame(split_count, split_count.index.tolist()).reset_index()
split_count.rename(columns={0: 'split_count'}, inplace=True)
split_count

plt.figure(figsize=(20,20))
plt.barh(split_count['index'], split_count['split_count'])

plt.show()
```

## Impurity

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.inspection import permutation_importance
import shap
from matplotlib import pyplot as plt
```

```python
rf = RandomForestRegressor(n_estimators=100)
rf.fit(X_train, y_train["Card number"].ravel())
```

```python
plt.figure(figsize=(20,20))
plt.barh(X.columns.tolist(), rf.feature_importances_)
```

```python
impurity=pd.DataFrame(rf.feature_importances_, X.columns.tolist()).reset_index()
impurity.rename(columns={0:'impurity'}, inplace=True)
```

```python
plt.figure(figsize=(20,20))
plt.barh(impurity['index'], impurity['impurity'])

plt.show()
```

##  F-statistic

```python
from sklearn.feature_selection import f_regression
f = pd.Series(f_regression(X, Y["Card number"].ravel())[0], index = X.columns)
f_stat = f / f.sum() * 100
```

```python
f_stat=pd.DataFrame(f_stat, f_stat.index.tolist()).reset_index()
f_stat.rename(columns={0: 'f_stat'}, inplace=True)
```

```python
plt.figure(figsize=(20,20))
plt.barh(f_stat['index'], f_stat['f_stat'])

plt.show()
```

## Maximal Information Coefficient

```python
from minepy import MINE
def get_mic(x, y):
    mine = MINE()
    mine.compute_score(x, y)
    return mine.mic()
f = X.apply(lambda X: get_mic(X, Y["Card number"]))
mic = f / f.sum() * 100
```

```python

mic=pd.DataFrame(mic, mic.index.tolist()).reset_index()
mic.rename(columns={0: 'mic'}, inplace=True)
```

```python
plt.figure(figsize=(20,20))
plt.barh(mic['index'], mic['mic'])

plt.show()
```

## Predictive Power Score

```python
import ppscore
f = ppscore.predictors(pd.concat([X, Y], axis = 1),
  'Card number').set_index('x')['ppscore']
pps = f / f.sum() * 100
```

```python
pps=pd.DataFrame(pps, pps.index.tolist()).reset_index()
```

```python
plt.figure(figsize=(20,20))
plt.barh(pps['index'], pps['ppscore'])

plt.show()
```

## Coverage

```python
from xgboost import XGBClassifier
f = pd.Series(xgb.get_booster().get_score(importance_type='cover'))
fimpo = f / f.sum() * 100
```

```python
coverage=pd.DataFrame(fimpo, fimpo.index.tolist()).reset_index()
coverage.rename(columns={0: 'coverage'}, inplace=True)
```

```python
plt.figure(figsize=(20, 20))
plt.barh(coverage['index'], coverage['coverage'])

plt.show()
```

## Permutation Importance

```python
from sklearn.inspection import permutation_importance
f = permutation_importance(rf, X, Y)['importances_mean']
fimpo = f / f.sum() * 100
```

```python
permutation=pd.DataFrame(fimpo, X.columns.tolist()).reset_index()
permutation.rename(columns={0: 'permutation'}, inplace=True)
```

```python
plt.figure(figsize=(20, 20))
plt.barh(permutation['index'], permutation['permutation'])

plt.show()
```

```python
import shap
explainer = shap.Explainer(rf)
shap_values = pd.DataFrame(explainer(X).values, columns = X.columns)
```

## Absolute Importance

```python
f = shap_values.abs().sum()
fimpo = f / f.sum() * 100
```

```python
absolute=pd.DataFrame(fimpo, fimpo.index.tolist()).reset_index()
absolute.rename(columns={0: 'absolute'}, inplace=True)
```

```python
plt.figure(figsize=(20,20))
plt.barh(absolute['index'], absolute['absolute'])

plt.show()
```

## Main Factor

```python
fimpo = shap_values.abs().idxmax(axis = 1).value_counts(normalize=True) * 100
```

```python
main_factor=pd.DataFrame(fimpo, fimpo.index.tolist()).reset_index()
main_factor.rename(columns={0: 'main_factor'}, inplace=True)
```

```python
plt.barh(main_factor['index'], main_factor['main_factor'])

plt.show()
```

## Main Factor Positive

```python
fimpo = shap_values.clip(lower = 0).abs().idxmax(axis = 1).value_counts(normalize = True) * 100
```

```python
main_factor_pos=pd.DataFrame(fimpo, fimpo.index.tolist()).reset_index()
main_factor_pos.rename(columns={0: 'main_factor_pos'}, inplace=True)
```

```python
plt.barh(main_factor_pos['index'], main_factor_pos['main_factor_pos'])

plt.show()
```

## Main Factor Negative

```python
fimpo = shap_values.clip(upper = 0).abs().idxmax(axis = 1).value_counts(normalize = True) * 100
```

```python
main_factor_neg=pd.DataFrame(fimpo, fimpo.index.tolist()).reset_index()
main_factor_neg.rename(columns={0: 'main_factor_neg'}, inplace=True)
```

```python
plt.barh(main_factor_neg['index'], main_factor_neg['main_factor_neg'])

plt.show()
```

# Feature importance table

```python
f=[pps, impurity, split_count, coverage, permutation, absolute, main_factor,
   main_factor_pos, main_factor_neg]

imp=pd.merge(f_stat, mic, on="index", how="outer").fillna(0)

for i in f:
  imp=pd.merge(imp, i, on="index", how="outer").fillna(0)
```

```python
imp.sort_values(['f_stat','mic','permutation','absolute','main_factor','coverage','ppscore',	
                  'impurity',	'split_count',	
                 ], ascending=[False, False, False
                                          ,False, False, False, False, False, False])
```

From the feature importance table we can find out that the important attributes are
* Playgrounds
* Ayurveda
* Drinking bars
* Gyms
* Opthalmologists
* M-Other Male Workers
* Literate Males

Out of the 7, 4 attributes makes sense to the problem statement, that is to find important features attributing to CVD disease
* **Drinking bars** -> place where there is more bars, people are more prone to get affected by CVD
* **M-other Male workers** -> These are the factory workers according to census and they seem to be more affected to CVD compared to others.
* **Literate males** -> More the males are literate, more they are prone to get into habits that cause CVD

The data collected is subjected to bias which seems to be the main cause for not very accurate results. But from the available results, we can clearly see that CVD is disease that affects males.
